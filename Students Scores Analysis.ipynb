# Drop unnamed column

```python
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

From the graph above, we have analysed that the number of females in the dataset is greater than that of males'.

# Parents' Education

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
gb
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.title("Realtionship Between Parents' Education and Students' Scores")
plt.show()
```

From the graph above, we have concluded that the education level of the parents has a significant impact on their child's scores.

# Parents' Marital Status

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("Realtionship Between Parents' Marital Status and Students' Scores")
plt.show()
```

From the graph above, we concluded that the marital status of the parents has negligible or no impact on their child's scores.

# Distribution of Ethnic Groups

```python
df["EthnicGroup"].unique()
```

```python
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()

l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupC["EthnicGroup"]]
print(mlist)

plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
